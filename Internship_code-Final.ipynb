```python
import pandas as pd
import random
import nltk
import string

# Function to tokenize a question
def tokenize(ques):
    ques = ques.lower()
    ques = nltk.word_tokenize(ques)
    return ques

# Function to replace tokens in a question
def replace(ques):
    t = random.choice(string.ascii_letters)
    if "{" and "}" in ques:
        for n, i in enumerate(ques):
            if ques.index("{") < n < ques.index("}"):
                if ques[n] in string.ascii_letters:
                    ques[n] = random.choice(string.ascii_letters)
                try:
                    if isinstance(int(i), int):
                        ques[n] = str(random.randint(-50, 50))
                except ValueError:
                    continue
            else:
                if ques[n] in string.ascii_letters:
                    ques[n] = str(t)
                try:
                    if isinstance(int(i), int):
                        ques[n] = str(random.randint(-50, 50))
                except ValueError:
                    continue
    else:
        for n, i in enumerate(ques):
            if ques[n] in string.ascii_letters:
                ques[n] = str(t)
            try:
                if isinstance(int(i), int):
                    ques[n] = str(random.randint(-50, 50))
            except ValueError:
                continue
    temp = ' '.join(ques)
    temp = temp.capitalize()
    return temp

# Read input questions from CSV files
input1 = pd.read_csv('input1.csv', header=None, delimiter=';', encoding='utf-8').values.tolist()
input2 = pd.read_csv('input2.csv', header=None, delimiter=';', encoding='utf-8').values.tolist()
input3 = pd.read_csv('input3.csv', header=None, delimiter=';', encoding='utf-8').values.tolist()

# Generate modified questions and write to Excel files
output_data1 = []
output_data2 = []
output_data3 = []

for row in input1:
    for _ in range(5):
        question = random.choice(row)
        modified_question = replace(tokenize(question))
        output_data1.append([modified_question])

for row in input2:
    for _ in range(5):
        question = random.choice(row)
        modified_question = replace(tokenize(question))
        output_data2.append([modified_question])

for row in input3:
    for _ in range(5):
        question = random.choice(row)
        modified_question = replace(tokenize(question))
        output_data3.append([modified_question])

output_df1 = pd.DataFrame(output_data1, columns=["Modified Question"])
output_df2 = pd.DataFrame(output_data2, columns=["Modified Question"])
output_df3 = pd.DataFrame(output_data3, columns=["Modified Question"])

output_df1.to_excel("output_questions1.xlsx", index=False, engine="openpyxl")
output_df2.to_excel("output_questions2.xlsx", index=False, engine="openpyxl")
output_df3.to_excel("output_questions3.xlsx", index=False, engine="openpyxl")
```
